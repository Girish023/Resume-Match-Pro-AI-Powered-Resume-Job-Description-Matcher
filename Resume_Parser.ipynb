```python
import spacy
from pdfplumber import open as pdf_open

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

def parse_resume(pdf_path):
    with pdf_open(pdf_path) as pdf:
        text = "\n".join([page.extract_text() for page in pdf.pages])
    doc = nlp(text)
    entities = {ent.label_: ent.text for ent in doc.ents}
    return entities
```

```python
def parse_job_description(job_text):
    doc = nlp(job_text)
    entities = {ent.label_: ent.text for ent in doc.ents}
    return entities
```

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

# Initialize model and FAISS index
model = SentenceTransformer('all-MiniLM-L6-v2')
index = faiss.IndexFlatL2(384)

def build_index(corpus):
    embeddings = model.encode(corpus)
    index.add(np.array(embeddings))

def search(query):
    query_embedding = model.encode([query])
    _, indices = index.search(np.array(query_embedding), k=5)
    return indices
```

```python
from flask import Flask, request, jsonify
from werkzeug.utils import secure_filename

app = Flask(__name__)

@app.route('/upload', methods=['POST'])
def upload_resume():
    file = request.files['resume']
    filename = secure_filename(file.filename)
    file.save(f"./uploads/{filename}")
    entities = parse_resume(f"./uploads/{filename}")
    return jsonify(entities)

if __name__ == '__main__':
    app.run(debug=True)
```

```python
import streamlit as st
import requests

st.title("Resume Job Matcher")

uploaded_file = st.file_uploader("Upload your resume", type=["pdf"])

if uploaded_file is not None:
    with open("temp_resume.pdf", "wb") as f:
        f.write(uploaded_file.getbuffer())
    response = requests.post("http://localhost:5000/upload", files={"resume": open("temp_resume.pdf", "rb")})
    st.write(response.json())
```
